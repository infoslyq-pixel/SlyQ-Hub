```python
from transformers import AutoProcessor, MusicgenForConditionalGeneration
import torch
import scipy.io.wavfile

# Model load
processor = AutoProcessor.from_pretrained("facebook/musicgen-small")
model = MusicgenForConditionalGeneration.from_pretrained("facebook/musicgen-small")

# Prompt (tune likh apni marzi ki)
prompt = ["a chill lo-fi beat with rain sounds"]

inputs = processor(
  text=prompt,
  padding=True,
  return_tensors="pt"
)

audio_values = model.generate(**inputs, max_new_tokens=256)

# Save output
scipy.io.wavfile.write("output.wav",
                       rate=model.config.audio_encoder.sampling_rate, data=audio_values[0].numpy())
print("‚úÖ AudioVerse: Music generated and saved as output.wav")
```
